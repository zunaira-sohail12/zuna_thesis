**Dropping Entitiy column**

```python
X = df.drop(['ENTITY'], axis=1)

y = df['ENTITY']
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

**Dropping Reference column from X train and test dataset**

```python
X_train.drop(['REFERENCES'], axis=1, inplace=True)
X_test.drop(['REFERENCES'], axis=1, inplace=True)
```

**Transform X train and test dataset**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
# import SVC classifier
from sklearn.svm import SVC


# import metrics to compute accuracy
from sklearn.metrics import accuracy_score


# instantiate classifier with default hyperparameters
svc=SVC()


# fit classifier to training set
svc.fit(X_train,y_train)


# make predictions on test set
y_pred=svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with default hyperparameters: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
# instantiate classifier with rbf kernel and C=100
svc=SVC(C=100.0)


# fit classifier to training set
svc.fit(X_train,y_train)


# make predictions on test set
y_pred=svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with rbf kernel and C=100.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

**Classification report for Y test and prediction test**

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_test))
```

```python
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
# print classification accuracy

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print('Classification accuracy : {0:0.4f}'.format(classification_accuracy))
```

```python
# print classification error

classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Classification error : {0:0.4f}'.format(classification_error))
```

```python
# print precision score

precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))
```

**Calculate Sensitivity**

```python
recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))
```

**Calculate Specifity**

```python
specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
```

```python
# Re-instantiate and fit linear_svc with probability=True to get probability estimates
from sklearn.svm import SVC
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_auc_score

# Instantiate classifier with linear kernel and C=1.0, and enable probability estimates
linear_svc_prob = SVC(kernel='linear', C=1.0, probability=True, random_state=0)

# Fit classifier to training set
linear_svc_prob.fit(X_train, y_train)

# Get probability estimates for the test set
y_pred_prob = linear_svc_prob.predict_proba(X_test)

# Ensure LabelBinarizer uses the same classes and order as the SVC model
# This is crucial for roc_auc_score with multi_class='ovr'
label_binarizer = LabelBinarizer()
label_binarizer.fit(linear_svc_prob.classes_) # Fit on the classes the model learned

y_test_binarized = label_binarizer.transform(y_test) # Now transform y_test

# Compute ROC AUC for multiclass classification using 'ovr' (one-vs-rest) strategy
ROC_AUC = roc_auc_score(y_test_binarized, y_pred_prob, multi_class='ovr')

print('ROC AUC (One-vs-Rest) : {:.4f}'.format(ROC_AUC))
```
