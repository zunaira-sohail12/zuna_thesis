**Bar plot for frequency of each Entity**

```python
ax = df.query('ENTITY != "Other"') \
    .groupby('ENTITY')['D1'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='FREQUENCY OF EACH ENTITY')
ax.set_xlabel('ENTITIES FREQUENCY')
plt.show()
```
